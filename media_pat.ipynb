```python
listao=[]
#separar e correr as cidades
for cidade in cidades:
    temp=df[df['Munic�pio']==cidade]
    
    #separar e correr as tecnologias
    for tecno in tecnologia:
        try:
            temp1=temp[temp['Tecnologia']==tecno]
            for x,y in zip(mes_orig,mes_velo):
                total_conexoes=temp1[x].sum()
                total_velo=temp1[y].sum()
                media=total_velo/total_conexoes
                grupo=temp1['Grupo'].unique()[0]
                tex=temp1['Tecnologia'].unique()[0]
                listao.append([grupo,cidade,x,tex,media])    
        except:
            pass
```

```python
fim=pd.DataFrame(listao, columns=['grupo','cidade','mes','tec','media'])
```

```python
fim=fim.dropna(how='any')
```
